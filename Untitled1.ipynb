```python
# 1. Variabel
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

# 2. Operasi Aritmetika
a = 10
b = 3

print("\nOPERASI ARITMETIKA")
print("Penjumlahan =", a + b)
print("Pengurangan =", a - b)
print("Perkalian =", a * b)
print("Pembagian =", a / b)
print("Pembagian Bulat =", a // b)
print("Modulus =", a % b)
print("Pangkat =", a ** b)

# 3. Operasi Perbandingan
print("\nOPERASI PERBANDINGAN")
print("a > b =", a > b)
print("a < b =", a < b)
print("a == b =", a == b)
print("a != b =", a != b)
print("a >= b =", a >= b)
print("a <= b =", a <= b)

# 4. Operasi Logika
x = True
y = False

print("\nOPERASI LOGIKA")
print("x and y =", x and y)
print("x or y =", x or y)
print("not x =", not x)

# 5. Percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi")
else:
    print("\nKondisi tidak terpenuhi")
```

```python
# 1. IF Sederhana
nilai = 85
print("Contoh IF Sederhana:")
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi.\n")

# 2. IF-ELSE
umur = 17
print("Contoh IF-ELSE:")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM.")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")

# 3. IF-ELIF-ELSE
hari = "Rabu"
print("Contoh IF-ELIF-ELSE:")
if hari == "Senin":
    print("Hari Senin - Saatnya kembali bekerja!")
elif hari == "Selasa":
    print("Hari Selasa - Jadwal rapat mingguan.")
elif hari == "Rabu":
    print("Hari Rabu - Ada diskon di beberapa toko.")
else:
    print("Hari lainnya - Atur jadwalmu dengan baik.\n")

# 4. IF Bersarang (Nested IF)
suhu = 35
print("Contoh IF Bersarang (Nested IF):")
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air.")
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan.")
else:
    print("Cuaca sepertinya cukup sejuk.\n")

# 5. Operasi Logika dalam IF
nilai_teori = 75
nilai_praktik = 80
print("Contoh IF dengan Operasi Logika:")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai.")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik.")
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori.")
else:
    print("Anda perlu meningkatkan nilai praktik.\n")

# 6. If Ternary
angka = -5
print("Contoh If Ternary:")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print("Angka =", angka, "=>", status)
```

```python
# 1. FOR Loop dengan range()
print("1) FOR loop dengan range()")
for i in range(5):
    print("Perulangan ke-", i)

print()

# 2. FOR Loop untuk list
print("2) FOR loop dengan list")
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah:
    print("Buah:", item)

print()

# 3. WHILE Loop
print("3) WHILE loop")
count = 0
while count < 5:
    print("count =", count)
    count += 1

print()

# 4. BREAK
print("4) BREAK")
for i in range(10):
    if i == 3:
        print("Loop dihentikan pada i =", i)
        break
    print("i =", i)

print()

# 5. CONTINUE
print("5) CONTINUE")
for i in range(5):
    if i == 2:
        print("Lewati i =", i)
        continue
    print("i =", i)

print()

# 6. NESTED LOOP
print("6) NESTED LOOP")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")

print()

# 7. ELSE pada loop
print("7) ELSE pada loop")
for x in range(3):
    print("x =", x)
else:
    print("Loop selesai tanpa break")

y = 0
while y < 3:
    print("y =", y)
    y += 1
else:
    print("While selesai tanpa break")

print()

# 8. PASS
print("8) PASS")
for i in range(3):
    if i == 1:
        pass
    print("i =", i)
```

```python
# Program cek bilangan

angka = int(input("Masukkan angka: "))

# Genap / Ganjil
if angka % 2 == 0:
    print("Bilangan Genap")
else:
    print("Bilangan Ganjil")

# Prima
if angka <= 1:
    print("Bukan bilangan prima")
else:
    prima = True
    for i in range(2, int(angka**0.5) + 1):
        if angka % i == 0:
            prima = False
            break

    if prima:
        print("Bilangan Prima")
    else:
        print("Bukan bilangan Prima")
```
